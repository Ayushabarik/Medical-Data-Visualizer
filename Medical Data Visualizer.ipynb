```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("medical_examination.csv")
```

```python
df['overweight']=(df['weight']/(df['height']/100)**2).apply
(lambda x : 1 if x > 25 else 0)
```

```python
df
```

```python
#Normalizing the data: 0 for 1(always good) and 1 for >1(always bad)
df['cholesterol']=df['cholesterol'].apply 
(lambda x : 0 if x==1 else 1)
```

```python
df
```

```python
df['gluc']=df['gluc'].apply 
(lambda x : 0 if x==1 else 1)
```

```python
df
```

```python
#cat plots using seaborn and melt function:

df_cat = pd.melt(frame = df, value_vars=['cholesterol','gluc','smoke','alco','active','overweight'], id_vars=['cardio'])
```

```python
graph = sns.catplot(x="variable", col="cardio", hue="value", data=df_cat, kind="count")
graph.set_axis_labels("variable", "total")
```

```python
#Heat Map

#Data cleaning:
df_heat=df[
        (df['ap_lo'] <= df['ap_hi']) & 
        (df['height'] >= df['height'].quantile(0.025)) &
        (df['height'] <= df['height'].quantile(0.975)) &
        (df['weight'] >= df['weight'].quantile(0.025)) &
        (df['weight'] <= df['weight'].quantile(0.975)) 
    ]    

#correlation matrix:
corr = df_heat.corr()

#Mask the upper triangle:
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, annot=True, fmt='.1f', mask=mask, vmin=.16, vmax=.32, center=0, square=True, linewidths=.5, cbar_kws={'shrink':.45, 'format':'%.2f'})
```
